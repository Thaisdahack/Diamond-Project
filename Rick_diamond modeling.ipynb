# Diamonds Project

## Importing libraries and datasets

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder


hist_data = pd.read_csv('diamonds.csv')
rick_data = pd.read_csv('rick_diamonds.csv')
```

## Data Clean

```python
hist_data.head(60)
```

```python
# treating categorical data (cut) -- import Label Encoder

number = LabelEncoder()
```

```python
hist_data['cut'] = number.fit_transform(hist_data['cut'].astype('str'))

rick_data['cut'] = number.fit_transform(rick_data['cut'].astype('str'))
```

```python
# treating categorical data (color) -- import Label Encoder
```

```python
hist_data['color'] = number.fit_transform(hist_data['color'].astype('str'))

rick_data['color'] = number.fit_transform(rick_data['color'].astype('str'))
```

```python
hist_data['color']
```

```python
# treating categorical data (clarity) -- 

hist_data['clarity'] = number.fit_transform(hist_data['clarity'].astype('str'))

rick_data['clarity'] = number.fit_transform(rick_data['clarity'].astype('str'))
```

## Testing models

```python
#separate variables
X = hist_data[['carat']]
y = hist_data['price']
```
